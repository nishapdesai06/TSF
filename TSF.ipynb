## Relationship exploration: Continuous Vs Continuous -- Scatter Charts
When the Target variable is continuous and the predictor is also continuous, we can visualize the relationship between the two variables using scatter plot and measure the strength of relation using pearson's correlation value.

```python
# Calculating correlation matrix
ContinuousCols=['Profit','Sales','Discount']

# Creating the correlation matrix
CorrelationData=DF[ContinuousCols].corr()
CorrelationData
```

```python
import seaborn as sns
sns.heatmap(CorrelationData,annot=True,cmap='Blues')
```

```python
ContinuousCols=['Sales','Discount']

# Plotting scatter chart for each predictor vs the target variable
for predictor in ContinuousCols:
    DF.plot.scatter(x=predictor, y='Profit', figsize=(10,5), title=predictor+" VS "+ 'Profit')
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(10,5))
ax.scatter(DF['Discount'],DF['Sales'])
ax.set_xlabel('Discount')
ax.set_ylabel('Sales')
plt.show()
```

```python
# Visual Inference using Grouped Bar charts
CategoricalColsList=['Ship Mode', 'Segment','Region','Sub-Category','Quantity']

import matplotlib.pyplot as plt
fig, PlotCanvas=plt.subplots(nrows=len(CategoricalColsList), ncols=1, figsize=(15,60))

# Creating Grouped bar plots for each categorical predictor against the Target Variable "Survived"
for CategoricalCol , i in zip(CategoricalColsList, range(len(CategoricalColsList))):
    CrossTabResult=pd.crosstab(index=DF[CategoricalCol], columns=DF['Category'])
    CrossTabResult.plot.bar(color=['red','green','blue'], ax=PlotCanvas[i], title=CategoricalCol+' Vs '+'Category')
```

```python
PS1=DF.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
PS1[:].plot.bar(color=['Pink','Blue'],figsize=(15,8))
plt.title('Profit/Loss & Sales across State')
plt.xlabel('State')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
PS2=DF.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
PS2[:].plot.bar(color=['Pink','Blue'],figsize=(10,8))
plt.title('Profit/Loss & Sales across Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
PS2=DF.groupby('Region')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
PS2[:].plot.bar(color=['Pink','Blue'],figsize=(15,8))
plt.title('Profit/Loss & Sales across Region')
plt.xlabel('Region')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
PS2=DF.groupby('Sub-Category')[['Profit','Discount']].sum().sort_values(by='Profit',ascending=False)
PS2[:].plot.bar(color=['Yellow','Green'],figsize=(15,10))
plt.title('Profit & Discount across Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Profit & Discount')
plt.show()
```
